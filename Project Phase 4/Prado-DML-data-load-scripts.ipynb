```python
import pandas as pd
filepath=r"C:\Products.csv"
data=pd.read_csv(filepath,sep=";")
data
```

```python
import cx_Oracle
myconnection = cx_Oracle.connect('hr/hr@//localhost:1521/orcl')

df2 = pd.DataFrame(data, columns= ['Product_id','Products','Price'])
for row in df2.itertuples():
    mycursor = myconnection.cursor()
    mycursor.execute("insert into products (product_id,product_name,price) values (:1,:2,:3)", (row[1], row[2], row[3]))
myconnection.commit()
```
